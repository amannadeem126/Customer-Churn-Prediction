```python
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
text_data = [
    "The service is very bad and disappointing",
    "I am happy with the internet speed",
    "Customer care is not helpful",
    "Very satisfied with the package",
    "Billing charges are too high"
]

df = pd.DataFrame(text_data, columns=["Customer_Feedback"])
df
```

```python
df["Tokens"] = df["Customer_Feedback"].apply(word_tokenize)
df
```
